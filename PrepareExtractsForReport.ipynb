```python
df = utils.load_1minute_fx_bars('EURUSD', 2009)[['close']]
df['return'] = df['close'].pct_change()
print tabulate.tabulate(df['2009-03-16'].head(8), headers=df.columns, tablefmt="pipe")
eurusd_ret_desc = df['return'].describe()
eurusd_ret = df['return']
eurusd = df['close']
```

```python
df = utils.load_1minute_fx_bars('USDJPY', 2009)[['close']]
df['return'] = df['close'].pct_change()
print tabulate.tabulate(df['2009-03-16'].head(8), headers=df.columns, tablefmt="pipe")
usdjpy_ret_desc = df['return'].describe()
usdjpy_ret = df['return']
usdjpy = df['close']
```

```python
df = utils.load_1minute_fx_bars('XAUUSD', 2009)[['close']]
df['return'] = df['close'].pct_change()
print tabulate.tabulate(df['2009-03-16'].head(8), headers=df.columns, tablefmt="pipe")
xauusd_ret_desc = df['return'].describe()
xauusd_ret = df['return']
xauusd = df['close']
```

```python
df = pd.DataFrame({'eurusd':eurusd_ret_desc, 'usdjpy': usdjpy_ret_desc, 'xauusd':xauusd_ret_desc})
print tabulate.tabulate(df, headers=df.columns, tablefmt='pipe')
```

```python
eurusd_ret.hist(bins=30, range=(-.002,.002),alpha=0.3)
usdjpy_ret.hist(bins=30, range=(-.002,.002),alpha=0.3)
xauusd_ret.hist(bins=30, range=(-.002,.002),alpha=0.3)
plt.title('histogram of closing price returns')
plt.savefig('reportresources/histo_of_closing_price-returns.png')
```

```python
fig, ax = plt.subplots(3,1, sharex=True)
eurusd.plot(ax=ax[0], label='EURUSD', style='b-')
usdjpy.plot(ax=ax[1], label='USDJPY', style='g-')
xauusd.plot(ax=ax[2], label='XAUUSD', style='r-')
ax[0].set_title('Closing Price')
ax[0].legend()
ax[1].legend()
ax[2].legend()

plt.savefig('reportresources/3-pairs-closing-price-subplots.png')
```

```python
fig, ax = plt.subplots(3,1, sharex=True)
(eurusd_ret+1).cumprod().plot(ax=ax[0], label='EURUSD', style='b-')
(usdjpy_ret+1).cumprod().plot(ax=ax[1], label='USDJPY', style='g-')
(xauusd_ret+1).cumprod().plot(ax=ax[2], label='XAUUSD', style='r-')
ax[0].set_title('Cumulative product of (1+return)')
ax[0].legend()
ax[1].legend()
ax[2].legend()
plt.savefig('reportresources/3-pairs-cumprod-returns.png')
```

```python
plt.figure()
(eurusd_ret+1).cumprod().plot(label='EURUSD')
(usdjpy_ret+1).cumprod().plot(label='USDJPY')
(xauusd_ret+1).cumprod().plot(label='XAUUSD')
plt.legend()
```
